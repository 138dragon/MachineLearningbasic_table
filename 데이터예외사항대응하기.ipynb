# 데이터 타입 조정
- 변수(칼럼) 타입 확인 : 데이터.dtypes
- 변수를 범주형으로 변경: 
    - 데이터['칼럼명'].astype('category')
- 변수를 수치형으로 변경:
    - 데이터['칼럼명'].astype('int')
    - 데이터['칼럼명'].astype('fload')
- NA 값의 처리
    - NA 갯수 체크 : 데이터.isna().sum()
    - NA 값 채우기: 데이터['칼럼명'].fillna(특정숫자)


```
import pandas as pd
```

```
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head() #세투사0 버지니카1 배시컬러2 즉, 품종은 범주형데이터 나머지는 숫자형데이터
```

```
#원핫인코딩
인코딩 = pd.get_dummies(아이리스)
인코딩.head() #원핫인코딩 안됨. 왜냐면 아이템이 숫자면 그냥 숫자로 인식함
```

```
print(아이리스.dtypes)
```

```
#품종 타입을 범주형으로 파꾸어 준다.
아이리스['품종'] = 아이리스['품종'].astype('category')
print(아이리스.dtypes) #품종이 카테고리 타입으로 바뀜
```

```
#원핫인코딩
인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```

------------------------------------------------ NA problem--------------------------------------------------

```
# NA값을 체크해 봅시다. (NA = NotAvailable 즉, 값이 없음)
아이리스.isna().sum()
```

```
아이리스.tail()
```

```
# NA값에 꽃잎폭 평균값을넣어주는 방법
mean = 아이리스['꽃잎폭'].mean()
아이리스['꽃잎폭'] = 아이리스['꽃잎폭'].fillna(mean)
아이리스.tail()
```

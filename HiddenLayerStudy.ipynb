```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 과거의 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
데이터 = pd.read_csv(파일경로)
데이터.head()
print(데이터.columns)
```

```
#독립변수, 종속변수 분리
독립 = 데이터[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = 데이터[['medv']]
print(독립.shape,종속.shape)
```

```
#모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
model.summary()
```

```
#데이터로 모델을 학습(fit)시킵니다.
model.fit(독립, 종속, epochs=500, verbose=0)
```

```
#loss확인
model.fit(독립, 종속, epochs=10)
```

```
#모델을 이용합니다.
model.predict(독립[0:5])
```

```
#실제 종속변수 값 살펴보기
종속[0:5]
```

아이리스 품종 분류

```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 과거의 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
print(아이리스.columns)
```

```
# 원 핫 인코딩 (숫자가 아닌 항목들을 0,1로 분류해 줌) ★★
아이리스 = pd.get_dummies(아이리스)
아이리스.head()
```

```
# 독립,종속변수 분리
독립 = 아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 아이리스[['품종_setosa','품종_versicolor','품종_virginica']]
print(독립.shape,종속.shape)
```

```
# 모델의 구조를 만듭니다.
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(5, activation='swish')(X)
H = tf.keras.layers.Dense(5, activation='swish')(H)
H = tf.keras.layers.Dense(5, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
```

```
# 모델을 학습시킵니다.
model.fit(독립, 종속, epochs=1000, verbose=0)
```

```
# 모델 확인
model.fit(독립, 종속, epochs=10)
```

```
# 모델을 이용합니다.
model.predict(독립[-5:])
```

```
# 기존 종속변수 확인
종속[-5:]
```
